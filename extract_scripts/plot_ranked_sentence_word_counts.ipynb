```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
enrich_corpus = 'wikipedia'
output_dir = './' + enrich_corpus + '/data/ranked_15000/'
metric = 'score'
```

```python
csv_path = output_dir + 'ranked-word-counts_' + metric + '.csv'
plot_path = output_dir + 'ranked-word-counts_' + metric + '.png'
```

```python
df_word_counts = pd.read_csv(csv_path)
```

```python
df_word_counts.max()
```

```python
x = df_word_counts[u'Unnamed: 0']


# Four axes, returned as a 2-d array

fig, axarr = plt.subplots(2, 2, figsize=(10,6))
fig.suptitle("Ranked Sentence Word Counts")

for ax in axarr.flatten():
    ax.set_yscale('log')
    ax.set_ylim([1,10**3])
    ax.set_xlabel('Sentence Index')
    ax.set_ylabel('Word Count')
    
axarr[0, 0].scatter(x, df_word_counts[u'en0'], s=1)
axarr[0, 0].set_title('English')
axarr[0, 1].scatter(x, df_word_counts[u'fr0'], s=1)
axarr[0, 1].set_title('French')
axarr[1, 0].scatter(x, df_word_counts[u'es0'], s=1)
axarr[1, 0].set_title('Spanish')
axarr[1, 1].scatter(x, df_word_counts[u'fi0'], s=1)
axarr[1, 1].set_title('Finnish')

fig.tight_layout()
plt.show()

fig.savefig(plot_path)
```
